```python
from PIL import Image
import os

TARGET_SIZE = (296, 296)  # 300 -> 296 (nearest multiple of 8)

folders = [
    '/content/sharp_5.5k',
    '/content/blur_5k',
    '/content/sharp_100_test',
    '/content/blur_100_test'
]

for folder in folders:
    files = os.listdir(folder)
    for fname in files:
        if fname.lower().endswith(('.png', '.jpg', '.jpeg')):
            path = os.path.join(folder, fname)
            img = Image.open(path).convert('RGB')
            if img.size != TARGET_SIZE:
                img = img.resize(TARGET_SIZE, Image.LANCZOS)
                img.save(path)
    print(f"✅ {folder}: {len(files)} images resized to {TARGET_SIZE}")

print("\n✅ All folders done!")
```

```python
import time, os, glob, threading

def early_stop_monitor():
    patience = 8          # wait 8 validations (8000 iters) of no improvement
    best_psnr = 0
    no_improve_count = 0
    last_seen_psnr = None
    log_pattern = "/content/drive/MyDrive/MLSP_Restormer/checkpoints/train_Restormer_Test_*.log"

    while True:
        time.sleep(300)  # check every 5 minutes (not 2)
        logs = sorted(glob.glob(log_pattern))
        if not logs:
            continue

        with open(logs[-1], 'r') as f:
            lines = f.readlines()

        # Only count actual validation lines from the log
        psnr_lines = [l for l in lines if 'Validation ValSet' in l or
                      ('psnr' in l.lower() and '# psnr' in l)]
        if not psnr_lines:
            continue

        try:
            latest_line = psnr_lines[-1]
            latest_psnr = float(latest_line.split('# psnr:')[-1].strip().split()[0])

            # Only act if we have a NEW validation result
            if latest_psnr == last_seen_psnr:
                continue  # same validation, skip

            last_seen_psnr = latest_psnr
            print(f"[Monitor] New validation PSNR: {latest_psnr:.4f} | Best: {best_psnr:.4f} | No-improve: {no_improve_count}/{patience}")

            if latest_psnr > best_psnr + 0.001:
                best_psnr = latest_psnr
                no_improve_count = 0
            else:
                no_improve_count += 1

            if no_improve_count >= patience:
                print(f"[Monitor] 🛑 Early stopping! No improvement for {patience} validations.")
                print(f"[Monitor] Best PSNR: {best_psnr:.4f}")
                os.system("kill -9 $(pgrep -f basicsr.train)")
                break
        except:
            continue

monitor_thread = threading.Thread(target=early_stop_monitor, daemon=True)
monitor_thread.start()
print("✅ Early stop monitor running — patience: 8 validations (8000 iters)")
```

```python
scheduler_path = "/content/Restormer/basicsr/models/lr_scheduler.py"

with open(scheduler_path, "r") as f:
    code = f.read()

# Fix get_lr to handle idx=None (happens when last_epoch == sum of all periods)
old = """    def get_lr(self):
        idx = get_position_from_periods(self.last_epoch,
                                        self.cumulative_period)
        current_weight = self.restart_weights[idx]"""

new = """    def get_lr(self):
        idx = get_position_from_periods(self.last_epoch,
                                        self.cumulative_period)
        if idx is None:
            idx = len(self.restart_weights) - 1
        current_weight = self.restart_weights[idx]"""

if old in code:
    code = code.replace(old, new)
    with open(scheduler_path, "w") as f:
        f.write(code)
    print("✅ lr_scheduler.py patched successfully")
else:
    print(" Pattern not found ")
```

```python
import os, glob, re, subprocess, threading, time, shutil

%cd /content/Restormer

# Auto-backup thread
def backup_loop():
    while True:
        time.sleep(600)
        src = "/content/Restormer/experiments/Restormer_Test"
        dst = "/content/drive/MyDrive/MLSP_Restormer/checkpoints"
        os.makedirs(dst, exist_ok=True)
        shutil.copytree(src, dst, dirs_exist_ok=True)
        print("✅ Auto-backup to Drive done")

t = threading.Thread(target=backup_loop, daemon=True)
t.start()

# Find latest state
states_dir = "/content/drive/MyDrive/MLSP_Restormer/checkpoints/training_states"
resume_path = None

if os.path.exists(states_dir):
    states = sorted(
        glob.glob(f"{states_dir}/*.state"),
        key=lambda x: int(os.path.basename(x).split(".")[0])
    )
    if states:
        resume_path = states[-1]
        print(f"Resuming from: {resume_path}")
    else:
        print("Starting fresh — no checkpoints found")
else:
    print("Starting fresh — checkpoint dir doesn't exist yet")

# Patch train.yml — ALL fixes in one place
with open("train.yml", "r") as f:
    yml = f.read()

# Resume state
if resume_path:
    yml = re.sub(r"resume_state:.*", f"resume_state: {resume_path}", yml)
else:
    yml = re.sub(r"resume_state:.*", "resume_state: ~", yml)

# All other fixes
yml = re.sub(r"total_iter:.*", "total_iter: 92000", yml)
yml = re.sub(r"periods:.*", "periods: [15000, 35001, 20999, 21000]", yml)
yml = re.sub(r"restart_weights:.*", "restart_weights: [1, 1, 0.5, 0.25]", yml)
yml = re.sub(r"eta_mins:.*", "eta_mins: [0.0003, 0.000001, 0.000001, 0.000001]", yml)
yml = re.sub(r"iters:.*", "iters: [92000]", yml)
yml = re.sub(r"batch_size_per_gpu:.*", "batch_size_per_gpu: 1", yml)
yml = re.sub(r"mini_batch_sizes:.*", "mini_batch_sizes: [1]", yml)
yml = re.sub(r"gt_size:.*", "gt_size: 192", yml)
yml = re.sub(r"gt_sizes:.*", "gt_sizes: [192]", yml)
yml = re.sub(r"mixup: false", "mixup: true", yml)

with open("train.yml", "w") as f:
    f.write(yml)

# Verify before training
print("\n=== train.yml key values ===")
for line in yml.split('\n'):
    if any(k in line for k in ['total_iter', 'periods', 'restart_weights',
                                 'gt_size:', 'batch_size', 'mixup:', 'resume_state']):
        print(line.strip())

print("\n🚀 Starting training...")
!python -m basicsr.train -opt train.yml
```
